```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_emails = pd.read_csv('combined_data.csv')
df_emails.head()
```

```python
# Word count (split by blanks) in spam/non-spam emails
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 4))
ax[0].set_xlim(left=0, right=20000)
ax[1].set_xlim(left=0, right=20000)
ax[0].set_title('Word count histogram - non-spam')
ax[1].set_title('Word count histogram - spam')
ax[0].set_ylim(0, 50000)
ax[1].set_ylim(0, 50000)
df_emails[df_emails['label'] == 0]['text'].map(lambda x: x.split(' ').__len__()).plot.hist(bins=1000, ax=ax[0])
df_emails[df_emails['label'] == 1]['text'].map(lambda x: x.split(' ').__len__()).plot.hist(bins=1000, ax=ax[1])
```
